## Descripcion del proyecto 

Uno de los problemas más comunes que enfrentan los gimnasios y otros servicios es la pérdida de clientes. ¿Cómo descubres si un cliente ya no está contigo? Puedes calcular la pérdida en función de las personas que se deshacen de sus cuentas o no renuevan sus contratos. Sin embargo, a veces no es obvio que un cliente se haya ido: puede que se vaya de puntillas.

Los indicadores de pérdida varían de un campo a otro. Si un usuario compra en una tienda en línea con poca frecuencia pero con regularidad, no se puede decir que ha huido. Pero si durante dos semanas no ha abierto un canal que se actualiza a diario, es motivo de preocupación: es posible que tu seguidor se haya aburrido y te haya abandonado.
En el caso de un gimnasio, tiene sentido decir que un cliente se ha ido si no viene durante un mes. Por supuesto, es posible que estén en Cancún y retomen sus visitas cuando regresen, pero ese no es un caso típico. Por lo general, si un cliente se une, viene varias veces y luego desaparece, es poco probable que regrese.
Con el fin de combatir la cancelación, Model Fitness ha digitalizado varios de sus perfiles de clientes. Tu tarea consiste en analizarlos y elaborar una estrategia de retención de clientes.

Tienes que:

- Aprender a predecir la probabilidad de pérdida (para el próximo mes) para cada cliente

- Elaborar retratos de usuarios típicos: selecciona los grupos más destacados y describe sus características principales

- Analizar los factores que más impactan la pérdida

- Sacar conclusiones básicas y elaborar recomendaciones para mejorar la atención al cliente:
1. Identificar a los grupos objetivo
2. Sugerir medidas para reducir la rotación
3. Describir cualquier otro patrón que observes con respecto a la interacción con los clientes

**El dataset incluye los siguientes campos**:

- 'Churn' — la cancelación para el mes en cuestión

- 'gender'
- 'Near_Location' — si el usuario vive o trabaja en el vecindario donde se encuentra el gimnasio
- 'Partner' — si el usuario es un trabajador de una compañía asociada (el gimnasio tiene empresas asociadas cuyos empleados obtienen descuentos; en esos casos el gimnasio almacena información sobre los empleadores de los clientes)
- romo_friends — si el usuario originalmente se inscribió mediante una oferta "trae a un amigo" (se utilizó el código promocional de un amigo cuando pagaron el primer abono)
- 'Phone' — si el usuario aportó el número de teléfono
- 'Age'
- 'Lifetime' — el tiempo (en meses) desde que el usuario llegó por primera vez al gimnasio

**Datos del registro de visitas y compras y datos sobre el estado actual de la membresía** 

- 'Contract_period' — 1 mes, 3 meses, 6 meses o 1 año
- 'Month_to_end_contract' — los meses que faltan hasta que expire el contrato
- 'Group_visits' — si el usuario participa en sesiones grupales
- 'Avg_class_frequency_total' — frecuencia media de visitas por semana a lo largo de la vida del cliente
- 'Avg_class_frequency_current_month' — frecuencia media de visitas por semana durante el mes en curso
- 'Avg_additional_charges_total' — cantidad total de dinero gastado en otros servicios del gimnasio: cafetería, productos deportivos, cosméticos, masajes, etc.

```python
#Funcion que separa la fecha por dia mes y año

def get_date(df): 
    columns = df.columns.tolist()
    idx = [columns.index(x) for x in columns if 'date' in x][0]
    
    df[columns[idx]] = pd.to_datetime(df[columns[idx]])
    df['month'] = df[columns[idx]].dt.month_name()
    df['year'] = df[columns[idx]].dt.year
```

```python
# Funcion para convertir todas las columnas y celdas a minuscula 


def convertir_a_minusculas(df):
    """
    Convierte todas las columnas de un DataFrame a minúsculas utilizando applymap y str.lower().
    También modifica los nombres de las columnas a minúsculas.
    
    Args:
    df (pandas.DataFrame): El DataFrame a convertir.
    
    Returns:
    pandas.DataFrame: El DataFrame con todas las columnas en minúsculas y con nombres de columna en minúsculas.
    """
    df = df.applymap(lambda s: s.lower() if type(s) == str else s)
    df.columns = df.columns.str.lower()
    return df
```

## 04. Construir un modelo para predecir la cancelación de usuarios

### Divide los datos en conjuntos de entrenamiento y validación utilizando la función train_test_split().

```python
# Divide los datos en características (la matriz X) y la variable objetivo (y)
y = df['churn']
X = df.drop(['churn'], axis=1) 

# Divide los datos en conjuntos de entrenamiento y prueba
# El 80% se utiliza para entrenamiento y el 20% para prueba
# Se fija una semilla aleatoria para reproducibilidad

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Estandariza los datos utilizando el StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)  # Ajusta el escalador utilizando los datos de entrenamiento

X_train_st = scaler.transform(X_train)  # Transforma los datos de entrenamiento

X_test_st = scaler.transform(X_test)  # Transforma los datos de prueba

# Imprime las primeras 5 filas de los datos de entrenamiento estandarizados

print(X_train_st[:5])

# Calcula la media de la primera columna de los datos de entrenamiento estandarizados

print('Media de los datos de entrenamiento estandarizados', np.mean(X_train_st[:, 0]))
```

### Entrena el modelo en el set de entrenamiento con dos métodos: regresión logística y  bosque aleatorio

```python
# Definición del algoritmo para el modelo

# Entrenar el modelo

lr_model = LogisticRegression(random_state=42)

# se   entrenada el modelo para hacer predicciones

lr_model.fit(X_train_st, y_train)
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)

# información sobre el modelo y las métricas de evaluación
print('Modelo: ', lr_model)
print('\tExactitud: {:.2f}'.format(accuracy_score(y_test, lr_predictions)))
print('\tPrecisión: {:.2f}'.format(precision_score(y_test, lr_predictions)))
print('\tRecall: {:.2f}'.format(recall_score(y_test, lr_predictions)))
```

```python
# Definición del algoritmo para el modelo

# Entrenar el modelo

rf_model = RandomForestClassifier(random_state=42)

# se  entrena el modelo para hacer predicciones

rf_model.fit(X_train_st, y_train)
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)

# información sobre el modelo y las métricas de evaluación

print('Model: ', rf_model)
print('\tAccuracy: {:.2f}'.format(accuracy_score(y_test, rf_predictions)))
print('\tPrecision: {:.2f}'.format(precision_score(y_test, rf_predictions)))
print('\tRecall: {:.2f}'.format(recall_score(y_test, rf_predictions)))
```

### Conclusion

    Se compararon dos modelos, Regresión Logística y Clasificador de Bosques Aleatorios, en el conjunto de entrenamiento.
    
1. Ambos modelos tuvieron una precisión similar, con valores de alrededor del 0.86.
2. Sin embargo, la Regresión Logística mostró un recall más alto que el Clasificador de Bosques Aleatorios, con una puntuación de 0.83 frente a 0.73, respectivamente.
3. La precisión se refiere a la proporción de predicciones correctas en la clase objetivo, donde la Regresión Logística obtuvo una puntuación de 0.86, mostrando un rendimiento ligeramente mejor.
4. El recall indica la capacidad de descubrir los objetos reales de la clase objetivo, y nuevamente la Regresión Logística superó al Clasificador de Bosques Aleatorios con una puntuación de 0.83 frente a 0.73.
5. En términos de exactitud, ambos modelos obtuvieron resultados similares, con puntajes cercanos al 0.91 para ambos.
6. En general, la Regresión Logística demostró un mejor rendimiento en términos de recall, lo que indica una mayor capacidad para identificar correctamente los objetos de la clase objetivo.

Por lo tanto, se puede concluir que el modelo de Regresión Logística brindó mejores resultados en esta comparación.

## 05. Crear clústeres de usuarios

```python
# Estandarizar los datos
sc = StandardScaler()
x_sc = sc.fit_transform(X)

# Construir una matriz de distancias basada en la matriz de características estandarizada
linked = linkage(x_sc, method='ward')

# Graficar un dendrograma
plt.figure(figsize=(15,10))
dendrogram(linked, orientation='top')
plt.show()

# Entrenar el modelo de agrupamiento con el algoritmo K-means y predecir los clusters de los clientes
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(x_sc)

# Calcular el coeficiente de silueta
sil_score = silhouette_score(x_sc, labels)

# Observar los valores promedio de las características para cada cluster
df['cluster_mean'] = labels

cluster_data = df.groupby('cluster_mean').mean()
display(cluster_data)
```

#### Conclusion

El dendrograma muestra que hay 5 grupos óptimos. Los grupos 2 y 4 tienen la media más alta para "near_location". El grupo 0 tiene la mayoría de los clientes asociados. El grupo 2 tiene la mayoría de los clientes con amigos de promoción. El grupo 1 contiene usuarios que no tienen un número de teléfono. La edad promedio en todos los grupos es casi la misma (alrededor de 29 años). Los clientes en el grupo 0 tienen el contrato más largo (más de 10 meses) en promedio y se quedan más tiempo que los demás, mientras que los clientes en el grupo 3 tienen el contrato más corto (alrededor de 2 meses). Los grupos con menor tasa de abandono son el 0 y el 4, y el género está equitativamente distribuido en estos grupos.

Los cargos adicionales promedio y la duración de la membresía para los grupos tienen una asimetría hacia la derecha. La frecuencia promedio de clases  se distribuye aproximadamente de manera normal en todos los grupos.

### Calcula la tasa de cancelación para cada clúster (utiliza el método groupby()). ¿Difieren en términos de tasa de cancelación? ¿Qué grupos son propensos a irse y cuáles son leales

```python
# ¿Difieren en términos de tasa de abandono? ¿Qué grupos son propensos a abandonar y cuáles son leales?

# Filtrar los registros donde churn es igual a 1 (abandono)
cluster_churn = df.query('churn == 1')

# Calcular la tasa de abandono para cada grupo
churn_rate = (cluster_churn.groupby('cluster_mean')['churn'].count() / df.groupby('cluster_mean')['churn'].count()) \
    .reset_index().rename(columns={'cluster_mean': 'clusters', 'churn': 'churn rate'})
churn_rate
```

Según la tabla anterior, las tasas de abandono difieren. Los grupos 3 y 4 tienen más probabilidades de abandonar, mientras que los grupos 1 y 2 son más leales.
